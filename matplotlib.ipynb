```python
x = np.random.rand(50)
y = np.random.rand(50)
## INTENSITY OF COLOUR IS HANDLED BY ALPHA PARAMETER
colours = np.random.rand(50)
sizes =100*np.random.rand(50)
plt.scatter(x,y , c = colours,s = sizes , alpha = 0.8)
plt.xlabel("x axis")
plt.ylabel(" y axis")
plt.title("this iis title")
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
## WE ARE DIRECTLY GIVING THE COLOUR HERE
colours = np.random.rand(50)
sizes =100*np.random.rand(50)
plt.scatter(x,y , c = 'b',s = sizes , alpha = 0.8)
plt.xlabel("x axis")
plt.ylabel(" y axis")
plt.title("this iis title")
```

```python
## MARKER PARAMETER
x = np.random.rand(50)
y = np.random.rand(50)
## WE ARE DIRECTLY GIVING THE COLOUR HERE
colours = np.random.rand(50)
sizes =100*np.random.rand(50)
plt.scatter(x,y , c = 'b',s = sizes , alpha = 0.8 , marker = 'x')
plt.xlabel("x axis")
plt.ylabel(" y axis")
plt.title("this iis title")
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
## WE ARE DIRECTLY GIVING THE COLOUR HERE
colours = np.random.rand(50)
sizes =100*np.random.rand(50)
plt.scatter(x,y , c = 'b',s = sizes , alpha = 0.8,marker = 'v')
plt.xlabel("x axis")
plt.ylabel(" y axis")
plt.title("this iis title")
```
